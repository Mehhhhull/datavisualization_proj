```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantative data.

Lets discuss some plots that allow us to visualize the distribution of a dataset. These plots are:                                                                                                                                            
->distplot/histplot

->jointplot

->pairplot

->rugplot

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()#this is qualative. only total_bill and tip is quantative
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=3,kde=True)#kernel density estimater
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=3,kde=True)#kernel density estimater
```

```python
sns.jointplot(x='total_bill', y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip',data=df,kind='hex')
```

```python
sns.pairplot(df,hue='size')
```

```python
sns.rugplot(df['tip'])
```
