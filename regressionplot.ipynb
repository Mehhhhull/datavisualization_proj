```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Implot()

regression plot is used in Ml so we will not fully discuss what,how and where it isused but  we will see how it is created and what are the values we can pt inside it.

```python
tips=sns.load_dataset("tips")
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],scatter_kws={'s':100})
```
